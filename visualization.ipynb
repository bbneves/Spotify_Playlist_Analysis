```python
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
playlist_df = pd.read_csv(f'./Resources/cleaned_playlists.csv', index_col=0)
```

```python
playlist_df
```

```python
songs_count_df = playlist_df.groupby(['artist_name'])['track_uri'].count().reset_index(name='count').sort_values(['count'], ascending=False)
```

```python
ten_or_more = songs_count_df[songs_count_df ['count'] >= 10]
```

```python
ten_or_more 
```
